```python
import pandas as pd

emp = pd.read_csv("data/employees.csv")

def change_date_format(date):

    # create a list of the substrings separated by '/',
    # so in your case ['DD', 'MM', 'YYYY'] 
    split_dt = date.split("/")

    return split_dt[2] + '-' + split_dt[0] + '-' + split_dt[1]   

emp.head()
```

```python
#apply this function to all elements of 'JOIN_DATE' columns
emp.loc[:, 'birth_date'] = emp.loc[:, 'birth_date'].apply(change_date_format)
emp.loc[:, 'hire_date'] = emp.loc[:, 'hire_date'].apply(change_date_format)
emp.head()
```

```python
emp.to_csv("data/employ.csv", index=False)
```
